```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()


# Load the Excel file
df = pd.read_excel("Sales_Data_Store.xlsx")

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# 📊 1. Total Sales Over Time
sales_by_date = df.groupby('Date')['Total Sale'].sum()
sales_by_date.plot(title='Total Sales Over Time', xlabel='Date', ylabel='Sales', figsize=(10,5))
plt.grid(True)
plt.show()

# Top 5 Products by Total Sale
top_products = df.groupby('Product')['Total Sale'].sum().sort_values(ascending=False)
top_products.plot(kind='bar', title='Top Selling Products', ylabel='Total Sale', figsize=(8,5), color='skyblue')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

#Sales by Category
sales_by_category = df.groupby('Category')['Total Sale'].sum()
sales_by_category.plot(kind='pie', autopct='%1.1f%%', title='Sales by Category', figsize=(6,6))
plt.ylabel('')
plt.show()

# Sales by Location
sales_by_location = df.groupby('Location')['Total Sale'].sum()
sales_by_location.plot(kind='bar', title='Sales by Location', ylabel='Total Sale', color='orange', figsize=(8,5))
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

top_customers = df.groupby('Customer ID')['Total Sale'].sum().sort_values(ascending=False).head(5)
top_customers.plot(kind='bar', title='Top 5 Customers', ylabel='Total Sale', color='green', figsize=(8,5))
plt.grid(True)
plt.show()
```
